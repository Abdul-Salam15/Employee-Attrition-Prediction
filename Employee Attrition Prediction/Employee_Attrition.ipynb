```python
# Target Variable(Attrition)
df['Attrition'] = df['Attrition'].replace({'No': 0, 'Yes':1})
```

```python
#encode binary variables
df['OverTime'] = df['OverTime'].map({'No': 0, 'Yes':1})
df['Gender'] = df['Gender'].map({'Female' : 0, 'Male' : 1})
df['Over18'] = df['Over18'].map({'Y' : 1, "M" : 0})
```

```python
df.head(1)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoding_columns = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
label_encoders = {}
for column in encoding_columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
df.head()
```

```python
X = df.drop(['Attrition', 'Over18'], axis = 1)
y = df['Attrition'].values
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
random_over = RandomOverSampler(random_state = 42)
X_over, y_over = random_over.fit_resample(X,y)
print(Counter(y_over))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_over, y_over, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve,roc_auc_score, classification_report
logistic_reg = LogisticRegression()
logistic_reg.fit(X_train, y_train)
prediction = logistic_reg.predict(X_test)
print("The Accuracy Score-> ", accuracy_score(y_test, prediction))
precision = precision_score(y_test, prediction)*100
recall = recall_score(y_test, prediction)*100
f1score = f1_score(y_test, prediction)*100
print(f"Precision Value-> {precision}%, Recall Value-> {recall}% F1 Score-> {f1score}%")
```

```python
fig = plt.figure(figsize = (12,6))
ax1 = fig.add_subplot(1,2,1)
cnf_matrix = confusion_matrix(y_test,prediction)
ax1 = sns.heatmap(pd.DataFrame(cnf_matrix), annot = True, cmap = 'Blues', fmt = 'd')
bottom, top = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
plt.xlabel('Predicted')
plt.ylabel('Expected')

ax2 = fig.add_subplot(1,2,2)
y_pred_proba = logistic_reg.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test,  prediction)
auc = roc_auc_score(y_test, prediction)
ax2 = plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
